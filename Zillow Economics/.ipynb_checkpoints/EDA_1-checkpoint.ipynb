Let's start with the 2016 data since it's fully populated. We'll read it into a pandas dataframe which will allow us to easily handle and explore all of this data

```python
train_df = pd.read_csv("data/train_2016_v2.csv", parse_dates=["transactiondate"])
```

```python
train_df.head(10) # We can use .head() to get the first 'N' items of the dataframe
```

Looks like there's 3 different columns: an ID, the transaction date, and the **logerror**. The **logerror** is our target value that we want to predict, defined as the log-error between the Zillow estimate and the actual sale price:

__logerror = log(Zestimate)âˆ’log(SalePrice)__

So in a nutshell we want to predict, given all of this data, how accurate Zillow's Zestimate system is at predicting the price of a given home with given data. From a business perspective, this would allow us to know whether or not for a particular home the Zestimate system will perform well and where it will perform poorly. This will allow us to, in the future, improve the Zestimate algorithm but focusing on the week points.

To begin our EDA let's take a look at the distribution of the logerror, to see how we should generally tackle this. 

```python
plt.figure(figsize=(8,6))
plt.scatter(range(train_df.shape[0]), np.sort(train_df.logerror.values))
plt.xlabel('index', fontsize=12)
plt.ylabel('logerror', fontsize=12)
plt.show()
```

Looks like the vast majority of the time, the error is very low being close to zero. That means that our model is doing decently well overall, but there's a number of outliers that seem to produce the most errors. These may warrant their own specific attention later.

We can't yet see the exact errors of the majority of the data (in the middle) since it's all very close to zero. Let's remove the outliers and then do a histogram plot so we can more clearly see how the majority is distributed:

```python
ulimit = np.percentile(train_df.logerror.values, 99)
llimit = np.percentile(train_df.logerror.values, 1)
train_df['logerror'].loc[train_df['logerror']>ulimit] = ulimit
train_df['logerror'].loc[train_df['logerror']<llimit] = llimit

plt.figure(figsize=(12,8))
sns.distplot(train_df.logerror.values, bins=50, kde=False)
plt.xlabel('logerror', fontsize=12)
plt.show()
```

Awesome! This is a nice clean normal distribution. Most of the errors are concentrated right around zero. Our goal will be to further tighten this graph (squash it inward from the sides) and eliminate those outliers!

Now let's explore the date field. Maybe people buy and sell houses around the same time of year. Let's check the number of transactions in each month.

As we could see from the data page as well

That being said, it does look like most of the purchases are concentrated near the middle. People might tend to purchase houses during the summer months then. More people on vacation, kids out of school, and generally a lower traffic / pressure time.

Finally, we'll take a quick look at the IDs. If there aren't many duplicates, then we know that most houses are only bought/sold once in this time period of 2016 to 2017.

### Properties 2016

Now that we have a high level view of our target, we can start looking at the data and features we have to work with. Taking a quick look at the head is always nice to get a quite idea of the type of data we are working with in each feature. 

```python
prop_df = pd.read_csv("data/properties_2016.csv")
prop_df.shape
prop_df.info()
```

Looks like there's lots floats, so we won't have to do much data transformation like we would have to do with categorical data. The data will take up 1.3+ GB, so we better be sure to have anough RAM available!

```python
prop_df.head(10)
```

This is why doing a `.head()` is always useful. Many of our columns have NaN values for the first 10 lines! That looks suspicious ... so let's first investigate those NaNs so they don't throw us off later. We'll check how many NaNs are in each column. 

```python
missing_df = prop_df.isnull().sum(axis=0).reset_index()
missing_df.columns = ['column_name', 'missing_count']
missing_df = missing_df.loc[missing_df['missing_count']>0]
missing_df = missing_df.sort_values(by='missing_count')

ind = np.arange(missing_df.shape[0])
width = 0.9
fig, ax = plt.subplots(figsize=(12,18))
rects = ax.barh(ind, missing_df.missing_count.values, color='blue')
ax.set_yticks(ind)
ax.set_yticklabels(missing_df.column_name.values, rotation='horizontal')
ax.set_xlabel("Count of missing values")
ax.set_title("Number of missing values in each column")
plt.show()
```

Lots of NaNs!

Since this was for the state of California, some of these are understandable. For example, many houses in California don't have basements since weather is quite warm, so it makes sense that we would have NaNs for the basementsqft. Not everyone will have a swimming pool, so anything related to swimming pools will have lots of NaNs. It's great to see that many useful ones are still very much in tact such as bedroomcnt, bathroomcnt, and roomcnt. 

Some of the more fine-grained features are missing too. For example, the square footage of each floor / section of the house is missing a lot, but the overall square footage isn't missing too much. Hopefully the total square footage is enough to encompass the value of the other more specific ones. 

Now that we know which features are mostly NULL and which are well populated, we can focus our EDA on the fuller ones, since features with many NaNs aren't really going to paint us a very accurate picture of our data anyways.

### Univariate Analysis

Before jumping in, let's merge our training and properties data

```python
train_df = pd.merge(train_df, prop_df, on='parcelid', how='left')
train_df.head()
```

Let's start by looking at only the 'float' variables alone and then get the correlation with the target logerror to see how they are related. This correlation analysis will give us a good idea of which variables will likely be the most effective in predicting our output. 

```python
# Let us just impute the missing values with mean values to compute correlation coefficients #
mean_values = train_df.mean(axis=0)
train_df_new = train_df.fillna(mean_values, inplace=True)

# Now let us look at the correlation coefficient of each of these variables #
x_cols = [col for col in train_df.columns if col not in ['logerror'] if train_df[col].dtype=='float64']

labels = []
values = []
for col in x_cols:
    labels.append(col)
    values.append(np.corrcoef(train_df[col].values, train_df.logerror.values)[0,1])
corr_df = pd.DataFrame({'col_labels':labels, 'corr_values':values})
corr_df = corr_df.sort_values(by='corr_values')
    
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(12,40))
rects = ax.barh(ind, np.array(corr_df.corr_values.values), color='y')
ax.set_yticks(ind)
ax.set_yticklabels(corr_df.col_labels.values, rotation='horizontal')
ax.set_xlabel("Correlation coefficient")
ax.set_title("Correlation coefficient of the variables")
#autolabel(rects)
plt.show()
```

Cool! There's a lot of rich information in the above graph.

There 4 items most correlated with our logerror are the square-footage of the home, number of bathrooms, number of bedrooms, and the year it was built. From a business perspective we can say that homes with more square footage, bathrooms, and bedrooms seem to yield higher error! This might make sense if we think about the fact that such homes would be more expensive, thus in general less sales, which also leaves us with less data to work with for homes in that price range! The year built is similar in that we will have less data for more recent sales simply due to time.

It seems that the city the house is in doesn't have much effect on the error of our Zestimate system, interestingly. BUT the county that the house is in have a strong negative correlation as do taxes.
